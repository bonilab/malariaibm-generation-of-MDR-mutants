# Fig.4 - Flow Diagrams for Mutation Pairs during Simulations

- 4 rows correspond to each 5-year period during the 20-year simulation (not counting first 10 year of burn-in).
- 3 columns are cycling, mft, and reversed-order cycling, of default PfPR/Cov settings.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

plt.rcParams['figure.figsize'] = [8.5*3, 11*3]
rc = {"font.family" : "sans-serif", 
      "font.style" : "normal",
      "mathtext.fontset" : "dejavusans"}
plt.rcParams.update(rc)
plt.rcParams["font.sans-serif"] = ["Myriad Pro"] + plt.rcParams["font.sans-serif"]
```

```python
# column 1, mft
r1c1 = mpimg.imread('fig4_set7m_5yr1.png')
r2c1 = mpimg.imread('fig4_set7m_5yr2.png')
r3c1 = mpimg.imread('fig4_set7m_5yr3.png')
r4c1 = mpimg.imread('fig4_set7m_5yr4.png')

# column 2, cyc
r1c2 = mpimg.imread('fig4_set7c_5yr1.png')
r2c2 = mpimg.imread('fig4_set7c_5yr2.png')
r3c2 = mpimg.imread('fig4_set7c_5yr3.png')
r4c2 = mpimg.imread('fig4_set7c_5yr4.png')

# column 3, reversed cyc
r1c3 = mpimg.imread('fig4_set7c_rev_5yr1.png')
r2c3 = mpimg.imread('fig4_set7c_rev_5yr2.png')
r3c3 = mpimg.imread('fig4_set7c_rev_5yr3.png')
r4c3 = mpimg.imread('fig4_set7c_rev_5yr4.png')
```
