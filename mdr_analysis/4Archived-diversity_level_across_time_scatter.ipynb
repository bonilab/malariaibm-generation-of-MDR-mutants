```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from constant import HEADER_NAME, COLUMNS_TO_DROP
```

```python
m7 = pd.read_csv('set7m_12monthly_data_0.txt', index_col=False, names=HEADER_NAME, sep='\t').drop(columns=COLUMNS_TO_DROP)
c7 = pd.read_csv('set7c_14monthly_data_0.txt', index_col=False, names=HEADER_NAME, sep='\t').drop(columns=COLUMNS_TO_DROP)
cr7 = pd.read_csv('set7cr_100monthly_data_0.txt', index_col=False, names=HEADER_NAME, sep='\t').drop(columns=COLUMNS_TO_DROP)
```

```python
def report_exceed_threshold_count(nparray, threshold):
  return np.count_nonzero(nparray >= threshold)
```

```python
het_calc_time = np.array(range(120,360,30))
```

```python
het_mft = []
het_cyc = []
het_cyc_rev = []
threshold = 0.001

for i in het_calc_time:
  het_mft.append(report_exceed_threshold_count(m7.iloc[i,:].values, threshold))
  het_cyc.append(report_exceed_threshold_count(c7.iloc[i,:].values, threshold))
  het_cyc_rev.append(report_exceed_threshold_count(cr7.iloc[i,:].values, threshold))
```

Plot

```python
plt.scatter(het_calc_time, het_mft, label='MFT')
plt.scatter(het_calc_time, het_cyc, label='5-Year Cycling')
plt.scatter(het_calc_time, het_cyc_rev, label='Reversed 5-Year Cycling')
plt.title('TBD 4Figb')
plt.xlabel('month')
plt.ylabel('Genotype count exceeding the threshold')
plt.legend()
```

```python
plt.bar(het_calc_time, het_mft, label='MFT', width = 5)
plt.bar(het_calc_time+5, het_cyc, label='5-Year Cycling', width = 5)
plt.bar(het_calc_time+10, het_cyc_rev, label='Reversed 5-Year Cycling', width = 5)
plt.title(f'TBD 4Figb, threshold={threshold}')
plt.xlabel('month')
plt.ylabel('Genotype count exceeding the threshold')
plt.legend()
```
